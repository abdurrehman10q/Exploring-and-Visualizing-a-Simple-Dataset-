```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

**1. Data Loading and Inspection using Pandas:**

This step involves importing datasets into Python using the pandas library, typically through functions like read_csv() or read_excel(). Once loaded, the data is inspected using methods such as .head(), .info(), and .describe() to understand the structure, data types, and basic characteristics of the dataset. This initial inspection helps in identifying missing values, data inconsistencies, and the overall shape of the dataset.

```python
#Loading The Datasets using pandas

data= pd.read_csv('/content/Iris.csv')

#printing the first few rows using .head()

data.head()
```

```python
data.shape #shows total number of rows and columns in dataset
```

**2. Descriptive Statistics and Data Exploration:**

Descriptive statistics provide a summary of the datasetâ€™s central tendency, dispersion, and distribution. This includes measures like mean, median, mode, standard deviation, and correlation coefficients. Through exploratory data analysis (EDA), insights are gathered by examining relationships, patterns, and anomalies in the data, which are crucial for guiding further analysis or model building.

```python
#Summary Statistics

data.describe()
```

```python
#use to check if there is null values in dataset
data.info()
```

**Visualizing The Data:**

**3. Basic Plotting and Visualization with Seaborn and Matplotlib:**

Data visualization is used to represent data graphically for better understanding and communication. Using matplotlib and seaborn, various plots such as histograms, scatter plots, and box plots are created. Seaborn, built on top of matplotlib, provides a higher-level interface for creating attractive and informative visualizations.

```python
#Scatter Plot
plt.figure(figsize=(8,8))
sns.scatterplot(data=data, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(8,8))
sns.scatterplot(data=data, x='PetalLengthCm', y='PetalWidthCm',hue="Species")
plt.show()
```

HISTOGRAMS

```python
#Histogram to Show Value Distribution

plt.figure(figsize=(5,5))
sns.histplot(data=data, x='SepalLengthCm', kde=False)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.histplot(data=data, x='SepalWidthCm', kde=False)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.histplot(data=data, x='PetalLengthCm', kde=False)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.histplot(data=data, x='PetalWidthCm', kde=False)
plt.show()
```

BOXPLOTS

```python
#Boxplot To Identify Outliers
plt.figure(figsize=(8,8))
sns.boxplot(data=data, x='SepalLengthCm',hue='Species')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=data, x='SepalWidthCm',hue='Species')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=data, x='PetalLengthCm',hue='Species')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=data, x='PetalWidthCm',hue='Species')
plt.show()
```
